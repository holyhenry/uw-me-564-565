# Simple Pendulum Example and Illustration

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
from pendulum_ode import *
```

```python
# define physical constants
g = -10
L = 10
d = 0.1

# define time slices and initial condition
t = np.linspace(0, 50, 500) 
y0 = [np.pi / 4, 0]

pendulum_sol = solve_ivp(pendulum_ode, (0, 50), y0, t_eval=t, args=(g, L, d,))
y = pendulum_sol.y
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(t, y[0, :])
plt.subplot(1, 2, 2)
plt.plot(y[0, :], y[1, :])

# make 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t, y[0, :], y[1, :])
plt.show()
```
