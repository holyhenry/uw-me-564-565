```python
# Exercise 1-2 (b)
t = np.linspace(0, 5, 1000)

f1 = 2 * np.exp(-5 * t)
f2 = 2 * np.exp(-1 * t)
f3 = [2] * 1000
f4 = 2 * np.exp(0.01 * t)
f5 = 2 * np.exp(0.1 * t)

plt.xlim(0, 5)
plt.ylim(-1, 5)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.plot(t, f1, 'b--', linewidth=1.2)
plt.plot(t, f2, 'r--', linewidth=1.2)
plt.plot(t, f3, 'g--', linewidth=1.2)
plt.plot(t, f4, 'm--', linewidth=1.2)
plt.plot(t, f5, 'c--', linewidth=1.2)
plt.legend(['λ1 = -5', 'λ2 = -1', 'λ3 = 0', 'λ4 = 0.01', 'λ5 = 0.1'])
plt.show()
```

```python
# Exercise 1-3 (a)
x = np.linspace(-5, 5, 2000)
fx1 = np.sin(x) / x
fx1_taylor = 1 - x**2/factorial(3) + x**4/factorial(5)
fx1_taylor4 = 1 - x**2/factorial(3) + x**4/factorial(5) - x**6/factorial(7)
fx1_taylor5 = 1 - x**2/factorial(3) + x**4/factorial(5) - x**6/factorial(7) + x**8/factorial(9)

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.plot(x, fx1, 'k', linewidth=1.5)
plt.plot(x, fx1_taylor, 'r--', linewidth=1.2)
# plt.plot(x, fx1_taylor4, 'g--', linewidth=1.0)
# plt.plot(x, fx1_taylor5, 'm--', linewidth=1.0)
plt.legend(['f(x) = sin(x)/x', 'taylor expansion(3rd)'])
# plt.legend(['f(x) = sin(x)/x', 'taylor expansion(3rd)', '(4th)', '(5rd)'])
plt.show()
```

```python
# Exercise 1-3 (b)
x = np.linspace(-5, 5, 2000)
fx2 = 3 ** x
fx2_taylor = 1 + np.log(3) * x + (np.log(3) * x) ** 2
fx2_taylor4 = 1 + np.log(3) * x + (np.log(3) * x) ** 2 / factorial(2) + (np.log(3) * x) ** 3 / factorial(3)
fx2_taylor5 = 1 + np.log(3) * x + (np.log(3) * x) ** 2 / factorial(2) + (np.log(3) * x) ** 3 / factorial(3) + (np.log(3) * x) ** 4 / factorial(4)

plt.xlim(-5, 5)
plt.ylim(-0, 5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.plot(x, fx2, 'k', linewidth=1.5)
plt.plot(x, fx2_taylor, 'r--', linewidth=1.2)
# plt.plot(x, fx2_taylor4, 'g--', linewidth=1.0)
# plt.plot(x, fx2_taylor5, 'm--', linewidth=1.0)
plt.legend(['f(x) = 3 ** x', 'taylor expansion(3rd)'])
# plt.legend(['f(x) = 3 ** x', 'taylor expansion(3rd)', '(4th)', '(5rd)'])
plt.show()
```

```python
# Exercise 1-4 (a)
t = np.arange(0, 10, 0.01, dtype=float)

ft1_real = np.cos(t) 
ft1_imag = np.sin(t)
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.grid(True)
plt.plot(ft1_real, ft1_imag, 'k', linewidth=1.2)
plt.legend(['f(t)'])
plt.show()
```

```python
# Exercise 1-4 (b)
t = np.arange(0, 10, 0.01, dtype=float)

ft2_real = np.exp(-t) * np.cos(t) 
ft2_imag = -np.exp(-t) * np.sin(t)
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.grid(True)
plt.plot(ft2_real, ft2_imag, 'k', linewidth=1.2)
plt.legend(['f(t)'])
plt.show()
```

```python
# Exercise 1-4 (c)
t = np.arange(0, 10, 0.01, dtype=float)

ft3_real = np.exp(1) * np.cos(t) 
ft3_imag = -np.exp(1) * np.sin(t)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.grid(True)
plt.plot(ft3_real, ft3_imag, 'k', linewidth=1.2)
plt.legend(['f(t)'])
plt.show()
```
